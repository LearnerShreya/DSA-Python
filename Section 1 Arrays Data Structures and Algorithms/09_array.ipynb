### **Big O of Common Array Operations (Time and Space Complexity)**  

---

#### 1. **Access**: **O(1) Time**, **O(1) Space**  
   - **Why?**  
     Direct access to an element at a specific index uses the formula:  
     \[  
     `Address} = Base Address + (Index * Size of Element)`  
     \]  
     This is constant-time and doesn’t consume additional memory.  

---

#### 2. **Set**: **O(1) Time**, **O(1) Space**  
   - **Why?**  
     Modifying an element at a specific index doesn’t involve traversal or additional memory.  

---

#### 3. **Traverse/Search**: **O(n) Time**, **O(1) Space**  
   - **Why?**  
     Traversal or search requires visiting each element once in the worst case. No extra space is needed since the operation is performed on the original array.  

---

#### 4. **Copy**: **O(n) Time**, **O(n) Space**  
   - **Why?**  
     Copying involves iterating through all \(n\) elements (time complexity) and storing them in a new array (space complexity).  

---

#### 5. **Insert**:  
   - **At the Beginning**: **O(n) Time**, **O(1) Space**  
     - **Why?** Shifting all elements to make space for the new element takes linear time. No extra space is required since the operation happens in place.  
   - **At the End**: **O(1) Time (Amortized)**, **O(1) Space**  
     - **Why?** If the array has room, appending is constant time. For dynamic arrays, resizing occasionally increases time but remains amortized \(O(1)\).  
   - **Somewhere in Between**: **O(n) Time**, **O(1) Space**  
     - **Why?** Shifting elements to make space for the new element takes linear time.  

---

#### 6. **Remove**:  
   - **At the Beginning**: **O(n) Time**, **O(1) Space**  
     - **Why?** All subsequent elements must be shifted to the left.  
   - **At the End**: **O(1) Time**, **O(1) Space**  
     - **Why?** Removing the last element requires no shifting or extra space.  
   - **Somewhere in Between**: **O(n) Time**, **O(1) Space**  
     - **Why?** Shifting elements after removal takes linear time.  

---

This analysis applies to **static arrays**. For **dynamic arrays**, additional space is required for resizing during insertions.

## Static & Dynamic array

#### **Static Array** 
- **Definition**: An array with a fixed size determined at the time of declaration. The size cannot change during runtime.
- **Characteristics**:
  - Fixed memory allocation.
  - More efficient in memory usage but less flexible.
  - Faster for access and manipulation since no resizing or reallocation occurs.
- **Languages**:  
  **C++**, **Java** (e.g., `int[]`), **C** (e.g., `int arr[10]`).

---

#### **Dynamic Array**
- **Definition**: An array that can grow or shrink during runtime as needed.  
- **Characteristics**:
  - Memory is dynamically allocated.
  - Can adjust size when elements are added or removed, often using **amortized \(O(1)\)** for appending.
  - Flexible but may involve overhead for resizing.
- **Languages**:  
  - **Python** (`list`): Built-in dynamic array.  
  - **JavaScript** (`Array`): Automatically adjusts size.  
  - **C++** (`std::vector`): A dynamic array class in the Standard Template Library (STL).  
  - **Java** (`ArrayList`): Provides a dynamic array implementation.

---

#### **Under the Hood Behavior**
1. **Python and JavaScript**:
   - Arrays are implemented as **dynamic arrays**. They resize when needed, typically doubling in capacity to minimize frequent reallocations.  
   - Example in Python:
     ```python
     arr = [1, 2, 3]  # Dynamic array
     arr.append(4)    # Expands as needed
     ```
   - Example in JavaScript:
     ```javascript
     let arr = [1, 2, 3];  // Dynamic array
     arr.push(4);          // Expands as needed
     ```

2. **C++ and Java**:
   - **Static Array**: 
     ```cpp
     int arr[5];  // Fixed size, cannot grow
     ```
     ```java
     int[] arr = new int[5];  // Static size
     ```
   - **Dynamic Array**:  
     - **C++**: Uses `std::vector`:
       ```cpp
       #include <vector>
       std::vector<int> arr = {1, 2, 3};
       arr.push_back(4);  // Grows dynamically
       ```
     - **Java**: Uses `ArrayList`:
       ```java
       import java.util.ArrayList;
       ArrayList<Integer> arr = new ArrayList<>();
       arr.add(4);  // Grows dynamically
       ```

---

#### **Key Points**  
- **Python/JavaScript**: Always dynamic arrays under the hood.  
- **C++/Java**: Provide both static and dynamic options (via `std::vector` or `ArrayList`).  
- Use static arrays for efficiency when size is known; use dynamic arrays for flexibility when size may vary.

### How can insertion have **O(1) space complexity** when extra memory is needed for a new element? Don’t we require one more block for insertion?

#### **Answer:**

Insertion in arrays is often said to have **O(1) space complexity**, but here's how this applies in different contexts:

---

#### **Static Arrays:**
1. **In-Place Modification**:  
   - When inserting into a static array, elements are **shifted within the same allocated memory space**.  
   - No new memory is allocated for the operation itself, so the space complexity is **O(1)**.

2. **No Resizing**:  
   - If the array is full, insertion is not possible unless you explicitly allocate a new array with larger size and copy elements, which requires **O(n)** additional memory.  
   - This resizing step is not part of a typical static array insertion process.

---

#### **Dynamic Arrays:**
1. **Enough Capacity**:
   - Dynamic arrays (e.g., Python's `list`, C++'s `std::vector`) internally manage extra capacity.
   - If capacity is available, the insertion happens directly within the pre-allocated memory → **O(1) space**.

2. **Resizing (When Capacity is Full)**:
   - If the array is full, resizing is triggered:
     - A **new memory block** is allocated with double the capacity.
     - All elements are copied to the new block.
     - This resizing step temporarily uses **O(n)** extra memory for copying elements.
   - Resizing happens infrequently, so the **amortized space complexity** of multiple insertions is still **O(1)**.

---

#### **Key Insights:**
- **Static Arrays**: Shifting elements doesn’t allocate new memory → **O(1) space**. But if resizing is needed, extra memory must be explicitly allocated.  
- **Dynamic Arrays**: Resizing involves a temporary **O(n)** cost, but the amortized space complexity for frequent insertions remains **O(1)** because resizing occurs only occasionally.
