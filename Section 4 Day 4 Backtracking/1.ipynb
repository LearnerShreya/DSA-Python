### **What is Backtracking?**
Backtracking is a controlled recursion.  
Backtracking is an algorithmic paradigm used to solve problems by exploring all possible solutions incrementally and abandoning ("backtracking") a partial solution as soon as it determines that it cannot lead to a valid or optimal answer. It efficiently prunes the search space by avoiding unnecessary explorations, making it suitable for solving problems with constraints.

---

### **How is Backtracking Different from Recursion?**

| **Aspect**           | **Recursion**                                | **Backtracking**                            |
|-----------------------|----------------------------------------------|---------------------------------------------|
| **Definition**        | A technique where a function calls itself to solve a smaller instance of a problem. | An extension of recursion where the algorithm explores all possibilities and backtracks upon reaching invalid states. |
| **Purpose**           | Focuses on dividing problems into subproblems. | Focuses on exploring all potential solutions under constraints. |
| **Constraint Handling** | Does not inherently handle constraints.     | Includes constraint validation at every step. |
| **Traversal**         | Typically, linear or depth-first traversal.  | Systematic exploration of decision trees with pruning. |

Backtracking **uses recursion** as its base but adds feasibility checks and pruning mechanisms.

---

### **How Does Backtracking Work?**
1. **Incremental Solution Construction**: Start with an empty solution and add elements one at a time.
2. **Feasibility Check**: At each step, check if the current solution is valid under the problem's constraints.
3. **Recursive Exploration**: If valid, recursively proceed to the next step; otherwise, discard the current choice (backtrack).
4. **Backtracking**: If a solution is not found after exploring all options at a particular step, backtrack by removing the last added element and trying alternative paths.
5. **Completion**: Stop when a complete solution is found, or all possibilities are exhausted.

---

### **Pass by Reference / Change Inplace**
In backtracking, solutions are often represented using mutable data structures like arrays or lists. By modifying these structures **inplace** (pass by reference), the algorithm avoids redundant memory allocation. After exploring a path, the changes are undone to restore the original state, ensuring correctness.

Example:
```python
def solve(board, row):
    if row == len(board):  # Base case: solution found
        print(board)
        return
    
    for col in range(len(board)):
        if is_valid(board, row, col):  # Check feasibility
            board[row][col] = 1  # Place queen (change inplace)
            solve(board, row + 1)  # Recursive call
            board[row][col] = 0  # Backtrack (undo change)
```

### **Blueprint to Solve Questions Using Backtracking**

1. **Understand the Problem**: Clearly define the problem, inputs, constraints, and outputs.
2. **Define the Recursive Function**:
   - **Base Case**: When a valid solution is constructed.
   - **Recursive Case**: Explore all valid options for the current step.
3. **Identify Constraints**: Incorporate checks to ensure feasibility at every step.
4. **Prune the Search Space**: Discard invalid paths early to reduce unnecessary exploration.
5. **Modify and Restore**: Change data inplace during recursion and backtrack to the original state after exploring a path.
6. **Test and Optimize**: Debug with smaller inputs and optimize pruning logic to improve performance.

---

### **Identify / When to Use Backtracking?**
Use backtracking when:
1. **Decision Tree Structure**: The problem can be represented as a tree of decisions, with each node representing a step.
2. **Constraints Exist**: The solution must meet specific conditions (e.g., no two queens threatening each other).
3. **Multiple Solutions Possible**: You need to explore all valid configurations or find an optimal one.
4. **Combinatorial Problems**: Problems involving permutations, combinations, or subsets (e.g., Sudoku, N-Queens, Graph Coloring).
5. **Brute Force Is Inefficient**: When brute force is too computationally expensive, and you can prune invalid paths early.
