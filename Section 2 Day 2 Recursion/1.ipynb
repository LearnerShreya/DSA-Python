Recursion is a programming technique where a function calls itself to solve smaller instances of the same problem. It breaks down complex problems into simpler subproblems, making them easier to solve. Here's a quick breakdown:

---

### **Key Concepts:**

1. **Base Condition**:
   - The stopping condition that prevents infinite recursion and ensures the function terminates. Without it, recursion would continue indefinitely, causing a stack overflow.
   - Example: `if n == 0: return 1`

2. **Recursive Case**:
   - The part of the function that breaks down the problem into smaller subproblems and makes the recursive call.
   - Example: `factorial(n) = n * factorial(n-1)`

3. **Recursion Tree**:
   - A visualization that shows the recursive calls and their relationships. Each node represents a function call.

4. **Recursion Call Stack**:
   - A structure that keeps track of function calls and their respective states, enabling backtracking and the unwinding process.

---

### **When to Use Recursion?**
- When a problem can be divided into smaller subproblems of the same type.
- Common in problems like factorial, Fibonacci sequence, tree traversals, and backtracking.

---
Recursive approach:
```python
def factorial(n):
    if n == 0:
        return 1  # Base case
    else:
        return n * factorial(n-1)  # Recursive case
```

Recursion is a powerful tool, but it requires careful handling of base conditions to avoid errors like infinite recursion and stack overflow.

---

#### **1. What is Recursion?**

   - **Recursion Tree**: A diagram showing all function calls and their relationships.

#### **4. Recursion vs Iteration**
- **Recursion**:
  - Great for problems with recursive patterns (e.g., tree traversals).
  - Higher memory usage (call stack).
- **Iteration**:
  - Efficient for problems like loops.
  - Uses less memory.

---

#### **5. Base Conditions**
- Base conditions prevent infinite recursion and define when the function should stop.
- Example: `if n == 0: return 1`
- **Importance**: Base conditions ensure recursion terminates correctly and avoids stack overflow.

---

#### **6. Recursive Leap of Faith**
- Trust that the recursive function will solve smaller instances of the problem correctly.

---

#### **7. Recurrence Relation**
- A mathematical expression that describes the recurrence of a problem.
- Example (Factorial):

# @25. Recursive Leap of Faith 13min 
-> how to use Recursion 
1) Understand the problem -> print 321123
2) Identify subproblem -> print 2112
3) Trust | Faith 
   - You have to trust that this recursive call will correctly solve a smaller version of your problem. 
   - You don't try to mentally simulate or "unroll" the recursion; 
   - instead, you assume that if your function works for simpler cases it will work for more complex ones.
4) Link 1 & 2 -> 3 + 2112 + 3
5) Base Condition 


```python
function seq(n):
{
    if n = 0 -> return
    print n 
    seq(n-1)   # inductive step
    print n
}

seq(3)
```

pseudocode:
```python
function seq(n):
    // Base condition: Stop recursion when n equals 0
    if n == 0:
        return
    
    // Recursive case: Print current number before recursive call
    print(n)
    
    // Recursive call to solve the subproblem
    seq(n - 1)
    
    // Recursive case: Print current number after recursive call
    print(n)

// Call the function with input
seq(3)
``` 

---
---
### Recursive Leap of Faith for 5!

---

### 1) Understand the Problem  
We want to calculate 5!, which is `5 * 4 * 3 * 2 * 1`.  

---

### 2) Identify the Subproblem  
To compute 5!, we need 4!, and for 4!, we need 3!, and so on until 1!.  
This defines the subproblem:  
`n! = n * (n-1)!`.

---

### 3) Trust | Faith  
- Trust that the recursive call will correctly compute `(n-1)!`.  
- Don't manually compute every step; assume that if your function works for 4!, 3!, etc., it will work for 5!.  

---

### 4) Link Subproblem to Solution  
Using the subproblem:  
- `5! = 5 * 4!`  
- `4! = 4 * 3!`  
- And so on, until the base case is reached.  

---

### 5) Base Condition  
The recursion stops when `n = 1`:  
- `1! = 1`.  
This is the base condition where recursion halts.  

---

### Pseudocode  

```python
function factorial(n):
    # Base Condition: Return 1 when n equals 1
    if n == 1:
        return 1
    
    # Recursive case: Multiply n by factorial of (n-1)
    return n * factorial(n - 1)

# Call the function
factorial(5)
```

---

### Recursive Call Stack for 5!  
1. `5! = 5 * factorial(4)`  
2. `4! = 4 * factorial(3)`  
3. `3! = 3 * factorial(2)`  
4. `2! = 2 * factorial(1)`  
5. Base case: `1! = 1`  

---

`5 * 4 * 3 * 2 * 1 = 120`
