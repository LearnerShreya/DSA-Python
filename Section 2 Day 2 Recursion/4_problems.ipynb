K-th Symbol in Grammar We build a table of n rows (1-indexed). We start by writing 0 in the 1st row. Now in every subsequent row, we look at the previous row and replace each occurrence of 0 with 01, and each occurrence of 1 with 10. For example, for n = 3, the 1st row is 0, the 2nd row is 01, and the 3rd row is 0110. Given two integer n and k, return the kth (1-indexed) symbol in the nth row of a table of n rows.

Clarifying Questions 
- Is it possible that n is given as 0? No, n>=1 
- Can k be out of bound? For eg. If n = 3, there will be 4 numbers, can k be given as 5 ? No, 1<=k<=2<sup>(n-1)</sup>

- **r1**: 0  
- **r2**: 0 1  
- **r3**: 0 1 1 0  
- **r4**: 0 1 1 0 1 0 0 1

Observation:
- 1) nth row, first half is same as previous row
- 2) nth row second half is NOT of previous row

````python

def kthGrammar(n, k):
    if n == 1 and k == 1:
        return 0
    
    parent = kthGrammar(n-1, (k+1)//2)
    if k % 2==1:
        return parent
    else:
        return 1-parent

print(kthGrammar(4,1))
print(kthGrammar(4,2))
print(kthGrammar(4,3))
print(kthGrammar(4,4))
print(kthGrammar(4,5))
print(kthGrammar(4,6))
print(kthGrammar(4,7))
print(kthGrammar(4,8))
        
````

1. From \( kthGrammar(1, 1) \): Parent value is \( 0 \).

````python
class Solution(object):
    def kthGrammar(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: int
        """
        # Base case: the first row only contains '0'
        if n == 1 and k == 1:
            return 0
        
        # Find the parent in the previous row
        parent = self.kthGrammar(n - 1, (k + 1) // 2)
        
        # Determine the value based on the parent's value and position
        if k % 2 == 1:  # Odd position, same as parent
            return parent
        else:  # Even position, opposite of parent
            return 1 - parent

# Create an instance of the Solution class
sol = Solution()

# Test the function
# Call the kthGrammar method
result = sol.kthGrammar(2, 2)

# Print the result
print(result)  # Output: 1
````

### Intuition
The problem is recursive in nature, as each row in the grammar is derived from the previous row. The K-th symbol in the N-th row can be determined by observing its "parent" symbol in the previous row. The key insight is that the relationship between rows follows a consistent rule:
- 0 produces 01
- 1 produces 10

This allows us to recursively calculate the symbol at any position without generating the entire table.

### Approach
1. **Base Case:**
   - For `n == 1` and `k == 1`, the result is `0`, as the first row only contains a single symbol `0`.

2. **Recursive Case:**
   - Find the "parent" of the K-th symbol in the (N-1)-th row. The position of the parent in the previous row is `(k + 1) // 2`.
   - Determine the value of the K-th symbol based on its parent:
     - If `k` is odd, the symbol is the same as its parent.
     - If `k` is even, the symbol is the opposite of its parent.

This approach avoids constructing the entire grammar table, relying instead on recursive calls to navigate the structure.

- **Space Complexity:**  
  The recursion stack requires space proportional to the depth of recursion, which is also \( O(n) \).

### Problem Description:
The function `kth_symbol` determines the value of the `k`-th symbol in the `n`-th row of a special sequence, based on the following rules:
1. The first row starts with `0`.
2. Each subsequent row is generated by replacing each `0` with `01` and each `1` with `10`.
   - Row 1: `0`
   - Row 2: `01`
   - Row 3: `0110`
   - Row 4: `01101001`, and so on.

The problem can be solved recursively because the `k`-th symbol in row `n` depends on its position in the row and the corresponding symbol in the previous row.

---

### Solution with Explanations:

```python
def kth_symbol(n, k):
    """
    Determines the k-th symbol in the n-th row of the sequence.
    
    Parameters:
    n (int): The row number (1-based index).
    k (int): The position in the row (1-based index).
    
    Returns:
    int: The value of the k-th symbol (0 or 1).
    """
    # Base case: The first row only contains '0'
    if n == 1:
        return 0
    
    length = 2 ** (n - 1)
    mid = length // 2  
    
    # Determine whether 'k' is in the first half or the second half
    if k <= mid:
        # If 'k' is in the first half, it mirrors the same position in the previous row
        return kth_symbol(n - 1, k)
    else:
        # If 'k' is in the second half, it's the opposite of the corresponding
        # position in the first half (k - mid maps to the first half)
        return 1 - kth_symbol(n - 1, k - mid)
        # return int(not kth_symbol(n-1, k-mid))
```

---

### Key Points:

1. **Base Case**:  
   - If `n == 1`, the first row always starts with `0`, so the answer is `0`.

2. **Recursive Case**:  
   - The length of row `n` is `2^(n-1)`. The midpoint (`mid`) divides the row into two equal halves.
   - If `k` lies in the first half (`k <= mid`), it is the same as the `k`-th symbol in the previous row (`n-1`).
   - If `k` lies in the second half (`k > mid`), the value is the opposite of the symbol in the corresponding position of the first half (`k - mid`).

3. **Opposite Calculation**:  
   - `1 - kth_symbol(...)` flips `0` to `1` and `1` to `0`.

---

### Example Usage:

```python
print(kth_symbol(4, 5))  # Output: 1
print(kth_symbol(3, 3))  # Output: 1
print(kth_symbol(2, 2))  # Output: 1
```

````python
# Time complexity = O(n^2) & SC = O(n)

def findTheWinner(n, k):
    arr = [i+1 for i in range(n)]
    def helper(arr, start_index):
        #base case 
        if len(arr)==1:
            return arr[0]
        #recursive case
        index_to_remove = (start_index + k - 1) % len(arr)
        del arr[index_to_remove]
        return helper(arr, index_to_remove)
    return helper(arr,0)  
````
