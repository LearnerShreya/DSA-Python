### **Practice Drawing a Recursion Tree**

A recursion tree helps visualize how recursive calls are made and solved. Let’s take the **Fibonacci series** as an example to illustrate this with pseudocode.

---

### **Fibonacci Series Recurrence Relation**
The recurrence relation for the Fibonacci series is:
\[
F(n) = F(n-1) + F(n-2)
\]
- Base cases:
  - \( F(0) = 0 \)
  - \( F(1) = 1 \)

---

### **Pseudocode**
```python
function Fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return Fibonacci(n-1) + Fibonacci(n-2)
```

---

### **Steps to Draw the Recursion Tree**
1. Start with the initial call \( F(n) \) as the root of the tree.
2. Break \( F(n) \) into its recursive calls \( F(n-1) \) and \( F(n-2) \).
3. Continue breaking down until reaching the base cases \( F(1) = 1 \) and \( F(0) = 0 \), which terminate the recursion.
4. Each level of the tree represents a deeper level of recursion.

---

### **Example: Recursion Tree for \( F(4) \)**

We want to compute \( F(4) \):
\[
F(4) = F(3) + F(2)
\]

#### **Pseudocode Execution**
```plaintext
Fibonacci(4)
    = Fibonacci(3) + Fibonacci(2)
    = (Fibonacci(2) + Fibonacci(1)) + (Fibonacci(1) + Fibonacci(0))
    = ((Fibonacci(1) + Fibonacci(0)) + 1) + (1 + 0)
```

---

### **Tree Representation**

```python
                 F(4)
               /     \
           F(3)       F(2)
          /   \       /   \
       F(2)   F(1)  F(1)  F(0)
      /   \
   F(1)   F(0)
```

---

### **Key Observations**
- Base cases (\( F(0) = 0 \) and \( F(1) = 1 \)) terminate recursion.
- Many subproblems are computed multiple times (e.g., \( F(2) \) and \( F(1) \) appear in multiple branches).

---

### **Benefits of Adding a Recursion Tree**
- **Visualization**: Helps understand how recursion unfolds and terminates.  
- **Debugging**: Spot logical errors by tracing calls.  
- **Optimization Insight**: Highlights redundancy (e.g., overlapping subproblems), motivating techniques like **memoization** or **dynamic programming**.

# @31. Recursion Approaches - O to N and N to 0 7min 

Recursive solutions can be implemented in two approaches:  
1. **0 to N**: Start from 0 and move towards \( N \).  
2. **N to 0**: Start from \( N \) and move towards 0.  

Practicing both approaches will help you build a strong grip on solving recursion-based problems.

---

### **Problem**  
**Given \( N \), calculate the sum of the series:**  
\[ 1 + 2 + 3 + ........ + N \]  

---

### **Approach 1: 0 to N**

**Recursive Function (Pseudocode)**:  
```python
function sum(curr, n):
    if curr == n:
        return n
    return curr + sum(curr + 1, n)

# Example call
result = sum(0, 5)
```

---

### **Step-by-Step Execution**:  

1. \( sum(0, 5) \): Current = 0 → Recursive call \( sum(1, 5) \).  
2. \( sum(1, 5) \): Current = 1 → Recursive call \( sum(2, 5) \).  
3. \( sum(2, 5) \): Current = 2 → Recursive call \( sum(3, 5) \).  
4. \( sum(3, 5) \): Current = 3 → Recursive call \( sum(4, 5) \).  
5. \( sum(4, 5) \): Current = 4 → Recursive call \( sum(5, 5) \).  
6. \( sum(5, 5) \): Base case → Return 5.  

---

### **Visualization**:  
```plaintext
sum(0, 5) → 15
    ↓ 
0 + sum(1, 5)
    ↓ ↑ 15
1 + sum(2, 5)
    ↓ ↑ 14
2 + sum(3, 5)
    ↓ ↑ 12
3 + sum(4, 5)
    ↓ ↑ 9
4 + sum(5, 5)
    ↓ 
    5
```

**Result**: \( 15 \)

---

### **Approach 2: N to 0**

**Recursive Function (Pseudocode)**:  
```python
function sum(n):
    if n == 0:
        return 0
    return n + sum(n - 1)

# Example call
result = sum(5)
```

---

### **Step-by-Step Execution**:  

1. \( sum(5) \): Current = 5 → Recursive call \( sum(4) \).  
2. \( sum(4) \): Current = 4 → Recursive call \( sum(3) \).  
3. \( sum(3) \): Current = 3 → Recursive call \( sum(2) \).  
4. \( sum(2) \): Current = 2 → Recursive call \( sum(1) \).  
5. \( sum(1) \): Current = 1 → Recursive call \( sum(0) \).  
6. \( sum(0) \): Base case → Return 0.  

---

### **Visualization**:  
```plaintext
sum(5) → 15
    ↓ 
5 + sum(4)
    ↓ ↑ 10
4 + sum(3)
    ↓ ↑ 6
3 + sum(2)
    ↓ ↑ 3
2 + sum(1)
    ↓ ↑ 1
1 + sum(0)
    ↓ 
    0
```

**Result**: \( 15 \)

---

### **Comparison**  
- **0 to N**: Easier to visualize when the problem builds up to \( N \).  
- **N to 0**: Common in problems involving reduction or breaking down large inputs.  

Both approaches are equally important for mastering recursion.

### **Parent of Key Techniques**
### **1. Backtracking**  
- **Definition**: A guided form of recursion used to explore all possible solutions systematically.  
- **Key Idea**: Recursively attempt to solve a problem, backtrack when a solution path is invalid, and try alternative paths.  
- **Example**: Solving mazes, N-Queens problem, generating permutations.

---

### **2. Dynamic Programming (DP)**  
- **Recursive Foundation**:  
  - Begin by writing the recursive solution.  
  - Transform it into a top-down approach with memoization (to store intermediate results).  
- **Top-Down (Memoization)**: A recursive approach optimized by storing results of already-solved subproblems.  
- **Bottom-Up (Tabulation)**: An iterative approach derived from the recursive solution. It builds the solution starting from the smallest subproblems.  

---

### **3. Greedy Algorithms**  
- **Definition**: A type of recursion where the focus is on optimizing the next immediate step rather than the overall solution.  
- **Key Idea**: Make the best local choice at each step, aiming for a globally optimal solution.  
- **Example**: Activity selection, Huffman encoding.

---

### **4. Divide and Conquer**  
- **Definition**: Breaks the problem into smaller, independent subproblems, solves each recursively, and combines the results.  
- **Key Idea**: Solutions to subproblems are combined to form the final solution.  
- **Example**: Merge Sort, Quick Sort, Binary Search.

---

### **When to Use Recursion**  
- **Tree or Branch-Based Problems**: If a problem requires exploring possibilities in a tree-like structure, recursion is often a natural fit.  
- **Dividing Problems into Subproblems**: If the problem can be divided into smaller, similar subproblems, recursion should be considered.  

--- 


# @33. Complexity Analysis of Recursive Solutions (2 min) 

#### **Key Formula for Time Complexity (T):**  
#### **T = Total Nodes × Work per Node**  

1. **Total Nodes:** The number of recursive calls made (size of the recursion tree).  
2. **Work per Node:** Computation done at each node (e.g., addition, multiplication).  

---

### Example: Fibonacci Recursion \( f(n) = f(n-1) + f(n-2) \)  

Let’s analyze \( f(4) \):  

#### **Recursion Tree**  
```plaintext
                f(4)
             /      \
          f(3)       f(2)
         /   \      /   \
      f(2)   f(1) f(1)  f(0)
     /   \
  f(1)  f(0)
```  

- **Nodes per level:**  
  - Level 0: 1 node (f(4))  
  - Level 1: 2 nodes (f(3), f(2))  
  - Level 2: 4 nodes (f(2), f(1), f(1), f(0))  
  - Level 3: 2 nodes (f(1), f(0))  

**Total Nodes:** Approximately 2ⁿ.  
**Work per Node:** O(1) (single addition).  

**Time Complexity:**  
T = O(2ⁿ)  

---

#### **Space Complexity:**  
- Maximum depth of recursion tree (call stack): O(n) for f(n).  

---

### General Approach to Analyze Recursive Solutions:  
1. **Visualize the Recursion Tree:**  
   - Identify the branching factor and levels.  
2. **Calculate Time Complexity:**  
   T = [Nodes per Level] × [Work per Node]  
3. **Calculate Space Complexity:**  
   Maximum depth of recursive calls (height of recursion tree).  

---

**Key Insight:**  
When dividing problems into subproblems or using recursion trees, consider:  
- Time Complexity (T):  
  **T = [Number of Nodes] × [Work Done per Node]**

- Space Complexity:  
  **Depth of the Recursive Call Stack** (determined by the maximum **depth of recursion**).
--- 

# @Quiz 5: Quiz: Recursion 
https://www.udemy.com/course/python-data-structures-algorithms/learn/quiz/6242892#content
