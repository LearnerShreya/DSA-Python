# @26. Visualising Recursion 9min 
### **Visualization**  

#### **1. Recursion Tree**  
- Represents how a recursive problem is divided into subproblems.  
- Each node: A function call.  
- Each branch: A smaller subproblem.  

**Example: Factorial of 3**  
```python
factorial(3)  
   └── 3 * factorial(2)  
         └── 2 * factorial(1)  
               └── 1  
```  
- Starts with the largest problem and breaks down to the base case.  

---

#### **2. Recursion Call Stack**  
- Tracks active recursive calls in memory.  
- **Push**: Adds a new call to the stack.  
- **Pop**: Removes a call when it completes.  

**Example: Call Stack for Factorial of 3**  
1. Factorial(3) is called → Push  
2. Factorial(2) is called → Push  
3. Factorial(1) is called → Push  
4. Factorial(1) returns 1 → Pop  
5. Factorial(2) returns 2 × 1 → Pop  
6. Factorial(3) returns 3 × 2 → Pop  

Final result: 3 × 2 × 1 = 6  

---

**Key Benefits of Visualization**  
- Simplifies complex recursion processes.  
- Enhances debugging and optimization.  

# @27. Recursion vs Iteration 4min 
- things done recurively can also be done iteratively
- does not use recursive call stack space [better space complexity]
- Recursion
  - ascending phase (calling phase)
  - descending phase (returning phase)
- iteration - only ascending phase

#### **1. Recursion**  
- A function calls itself to solve a problem.  
- Suitable for problems that naturally follow a divide-and-conquer or tree-like structure.  

**Advantages**  
- Cleaner and more readable code for complex problems (e.g., tree traversals).  
- Easier to implement solutions involving backtracking.  

**Disadvantages**  
- Higher memory usage due to the recursion call stack.  
- Risk of stack overflow for deep recursions without a proper base condition.  

---

#### **2. Iteration**  
- A loop is used to repeatedly execute a block of code.  
- Suitable for problems with a fixed, predictable number of steps.  

**Advantages**  
- More memory-efficient since no additional call stack is required.  
- Better performance for problems without a recursive nature.  

**Disadvantages**  
- Code can be less readable and more complex for some problems.  
- Difficult to implement solutions like backtracking using iteration.  

---

#### **Key Differences**  

| Feature          | Recursion                  | Iteration               |  
|-------------------|----------------------------|--------------------------|  
| **Structure**    | Function calls itself      | Uses loops (e.g., for, while) |  
| **Memory Usage** | Higher (stack memory)      | Lower                   |  
| **Complexity**   | Simple for recursive tasks | Complex for recursive tasks |  
| **Risk**         | Stack overflow possible    | No stack overflow risk  |  
| **Performance**  | Slower for deep recursions | Faster due to no overhead |  

---

#### **Example**: Factorial of \( n \)  

**Recursive Approach**  
```python
def factorial_recursive(n):
    if n == 1:
        return 1
    return n * factorial_recursive(n - 1)
```  

**Iterative Approach**  
```python
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```  

Both approaches yield the same result, but their efficiency and readability depend on the problem's nature.  

# @28. Ways to write Base condition 4min 
- A recursion is a function that class itself again and again untill a base condition or terminating condition occur.
- Base condition: tells when to stop recursively calling itself and returning
- Ways to write Base condition is to identify:
  - last valid input
  - first invalid input

---

#### **1. Last Valid Input**  
The base condition is triggered when the input reaches its last valid value.

**Example: Factorial Calculation**  
For \( n! \), the last valid input is \( n = 1 \). The recursion stops here.  
```python
def factorial(n):
    if n == 1:  # Last valid input
        return 1
    return n * factorial(n - 1)

factorial(5)  # Calculates 5! = 5 × 4 × 3 × 2 × 1
```

---

#### **2. First Invalid Input**  
The base condition is triggered when the input becomes invalid or goes out of range.

**Example: Countdown Printing**  
For a countdown from \( n \), the first invalid input is \( n = 0 \). The recursion stops here.  
```python
def countdown(n):
    if n == 0:  # First invalid input
        return
    print(n)
    countdown(n - 1)

countdown(5)  # Prints: 5, 4, 3, 2, 1
```

---

### **Key Differences**
| Type                  | Trigger Point            | Use Case Example         |  
|-----------------------|--------------------------|---------------------------|  
| **Last Valid Input**  | When input reaches its smallest valid value | Factorial, Fibonacci   |  
| **First Invalid Input** | When input exceeds its valid range          | Countdown, Array traversal |  

Both approaches ensure recursion terminates and prevent infinite function calls.

### **Recurrence Relation**

A **recurrence relation** is a way to describe a problem by expressing its solution in terms of smaller subproblems. It is often used in recursive algorithms to break down complex problems.

---

### **Structure of a Recurrence Relation**
1. **Recursive Case**: Defines the problem in terms of smaller instances.  
2. **Base Condition**: Specifies the stopping point of the recursion (smallest subproblem).  

---

### **Examples**

#### **1. Factorial of Positive Integers**
- **Recurrence Relation**:  
  \( F(n) = n \cdot F(n-1) \)  
  **Base Condition**: \( F(0) = 1 \)

- **Explanation**:  
  To compute \( n! \):  
  - Multiply \( n \) by the factorial of \( (n-1) \).  
  - Stop when \( n = 0 \).

- **Example for \( 5! \)**:  
  Step-by-step computation:  
  - \( 5! = 5 \cdot 4! \)  
  - \( 4! = 4 \cdot 3! \)  
  - \( 3! = 3 \cdot 2! \)  
  - \( 2! = 2 \cdot 1! \)  
  - \( 1! = 1 \)  

  Final result: \( 5! = 120 \).

---

#### **2. Fibonacci Sequence**
- **Recurrence Relation**:  
  \( F(n) = F(n-1) + F(n-2) \)  
  **Base Conditions**:  
  \( F(0) = 0, \, F(1) = 1 \)

- **Explanation**:  
  Each Fibonacci number is the sum of the two preceding numbers.

- **Example for \( F(4) \)**:  
  Step-by-step computation:  
  - \( F(4) = F(3) + F(2) \)  
  - \( F(3) = F(2) + F(1) \)  
  - \( F(2) = F(1) + F(0) \)  

  Final result: \( F(4) = 3 \).

---

#### **3. Printing Sequence 321123**
- **Recurrence Relation**:  
  - Print \( n \).  
  - If \( n > 0 \), recursively call \( F(n-1) \).  
  - Print \( n \) again.  

- **Pseudocode**:  
  ```python
  def F(n):
      if n == 0:
          return
      print(n)
      F(n-1)
      print(n)

  # Example call:
  F(3)
  ```

- **Output**:  
  ```
  3
  2
  1
  1
  2
  3
  ```

---

### **Key Points**
- Recurrence relations express a problem in terms of smaller subproblems.  
- A well-defined **base condition** is essential to avoid infinite recursion.  
- Recurrence relations are widely used in recursion, dynamic programming, and divide-and-conquer algorithms.
