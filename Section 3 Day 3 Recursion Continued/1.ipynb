````python
def toh(N, fromm, to, aux):
    count = 0

    def helper(N, fromm, to, aux):
        nonlocal count
        # Base case
        if N == 1:
            print("move disk " + str(N) + " from rod " + str(fromm) + " to rod " + str(to))
            count += 1
            return
        # Recursive case
        helper(N - 1, fromm, aux, to)
        print("move disk " + str(N) + " from rod " + str(fromm) + " to rod " + str(to))
        count += 1
        helper(N - 1, aux, to, fromm)

    helper(N, fromm, to, aux)
    return count
````

````python
# Example for N = 3
moves = toh(1, 1, 3, 2)
print("Total moves:", moves)
````

````python
# Example for N = 3
moves = toh(3, 1, 3, 2)
print("Total moves:", moves)
````

````python
def TowerOfHanoi(n, source, destination, auxiliary, move_count):
    if n == 1:
        print(f"Move {move_count}: Move disk 1 from {source} to {destination}")
        return move_count + 1

    # Step 1: Move n-1 disks from source to auxiliary
    move_count = TowerOfHanoi(n - 1, source, auxiliary, destination, move_count)
    # Step 2: Move the nth disk from source to destination
    print(f"Move {move_count}: Move disk {n} from {source} to {destination}")
    move_count += 1
    # Step 3: Move n-1 disks from auxiliary to destination
    move_count = TowerOfHanoi(n - 1, auxiliary, destination, source, move_count)

    return move_count

# Example Usage
n = 3
print(f"Solving Tower of Hanoi for {n} disks:")
total_moves = TowerOfHanoi(n, 'A', 'C', 'B', 1)
print(f"Total moves: {total_moves - 1}")
````
