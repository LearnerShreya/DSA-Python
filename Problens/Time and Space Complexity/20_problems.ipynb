### **1. Find the maximum and minimum element in an array**

```python
def find_max_min(arr):
    max_val = arr[0]
    min_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
    return max_val, min_val
```
- **Time Complexity (T):** **O(n)**
  - We go through each element in the array once to compare with the current max and min. The time complexity is linear in the size of the array.
- **Space Complexity (S):** **O(1)**
  - We only use a constant amount of extra space (for `max_val` and `min_val`), so the space complexity is constant.

### **2. Reverse an array in place**

```python
def reverse_array(arr):
    left, right = 0, len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr
```
- **Time Complexity (T):** **O(n)**
  - The array is traversed once, with the left and right pointers moving towards each other. Each element is swapped once, resulting in O(n) time.
- **Space Complexity (S):** **O(1)**
  - We only use two extra variables (`left` and `right`), so the space complexity is constant.

### **3. Rotate an array by `k` positions (in place)**

```python
def rotate_array(arr, k):
    n = len(arr)
    k = k % n  # Handle cases where k > n
    arr.reverse()
    arr[:k] = reversed(arr[:k])
    arr[k:] = reversed(arr[k:])
    return arr
```
- **Time Complexity (T):** **O(n)**
  - Reversing the array and slicing takes O(n) time.
- **Space Complexity (S):** **O(1)**
  - No additional space is used except for the in-place modifications to the array.

### **4. Implement binary search in a sorted array**

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # target not found
```
- **Time Complexity (T):** **O(log n)**
  - Each time, we reduce the search space by half, which results in logarithmic time complexity.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of space is used for variables (`left`, `right`, `mid`).

### **5. Search for a target in a rotated sorted array**

```python
def search_rotated(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        if arr[left] <= arr[mid]:  # Left side is sorted
            if arr[left] <= target < arr[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:  # Right side is sorted
            if arr[mid] < target <= arr[right]:
                left = mid + 1
            else:
                right = mid - 1
    return -1  # target not found
```
- **Time Complexity (T):** **O(log n)**
  - The array is still divided in half with each iteration, so the time complexity is logarithmic, even though it is rotated.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of space is used for variables.

### **6. Find the first occurrence of a target in a sorted array**

```python
def find_first_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            result = mid
            right = mid - 1  # Keep searching on the left side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result
```
- **Time Complexity (T):** **O(log n)**
  - Binary search is used to find the target, so the time complexity is logarithmic.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of space is used for variables.

### **7. Implement bubble sort**

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```
- **Time Complexity (T):** **O(n²)**
  - In the worst case, we compare every pair of elements, so it is quadratic.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of extra space is used for temporary variables.

### **8. Implement selection sort**

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```
- **Time Complexity (T):** **O(n²)**
  - The outer loop runs n times, and for each iteration, the inner loop runs n-i times.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of space is used for variables.

### **9. Implement insertion sort**

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```
- **Time Complexity (T):** **O(n²)**
  - The inner while loop compares elements in a sorted subset, and in the worst case, it runs O(n²) comparisons.
- **Space Complexity (S):** **O(1)**
  - Only a constant amount of space is used for temporary variables.

### **10. Implement merge sort**

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr
```
- **Time Complexity (T):** **O(n log n)**
  - The array is repeatedly split in half (log n times), and each merge operation takes linear time.
- **Space Complexity (S):** **O(n)**
  - We need additional space to store the temporary left and right sub-arrays.

### **11. Reverse a string**

```python
def reverse_string(s):
    return s[::-1]
```
- **Time Complexity (T):** **O(n)**
  - We traverse through the string once to reverse it.
- **Space Complexity (S):** **O(n)**
  - We create a new string that is the reverse of the input, so extra space proportional to the string length is used.

### **12. Check if a string is a palindrome**

```python
def is_palindrome(s):
    return s == s[::-1]
```
- **Time Complexity (T):** **O(n)**
  - The string is compared with its reverse, which takes linear time.
- **Space Complexity (S):** **O(n)**
  - A new reversed string is created, using extra space proportional to the string length.

### **13. Find the first non-repeating character in a string**

```python
from collections import Counter

def first_non_repeating(s):
    counts = Counter(s)
    for char in s:
        if counts[char] == 1:
            return char
    return None
```
- **Time Complexity (T):** **O(n)**
  - We iterate through the string twice: once to build the frequency map and once to find the first non-repeating character.
- **Space Complexity (S):** **O(n)**
  - We use a hash map to store character frequencies, which takes O(n) space.
