### **Understanding Recursion Unwinding:**

When you call a recursive function, it keeps **breaking down the problem** into smaller subproblems until it reaches the **base case**, where it stops further recursion.  

Once the base case is reached, the function starts **returning values back step-by-step**, which is called **unwinding**.  

---

### **Example: Sum of Digits (n = 568345)**  

Let’s take a step-by-step example with the function:

```python
def sum_of_digits(n):
    if n == 0:  
        return 0
    return (n % 10) + sum_of_digits(n // 10)
```

#### **Step 1: Breaking Down (Forward Process)**

If we call `sum_of_digits(568345)`, the function breaks it down into smaller parts:  

```
sum_of_digits(568345) → 5 + sum_of_digits(56834)
sum_of_digits(56834)  → 4 + sum_of_digits(5683)
sum_of_digits(5683)   → 3 + sum_of_digits(568)
sum_of_digits(568)    → 8 + sum_of_digits(56)
sum_of_digits(56)     → 6 + sum_of_digits(5)
sum_of_digits(5)      → 5 + sum_of_digits(0)
sum_of_digits(0)      → 0 (Base case reached, stop recursion)
```

At this point, recursion stops because `n == 0`.

---

#### **Step 2: Unwinding (Backward Process)**

Now the function **returns values step by step**, from the deepest call back to the top.  

We now **return the results** starting from the base case:

```
sum_of_digits(0)    → returns 0
sum_of_digits(5)    → 5 + 0 = 5
sum_of_digits(56)   → 6 + 5 = 11
sum_of_digits(568)  → 8 + 11 = 19
sum_of_digits(5683) → 3 + 19 = 22
sum_of_digits(56834)→ 4 + 22 = 26
sum_of_digits(568345)→ 5 + 26 = 31
```

---

#### **Final Answer:**  
When the function unwinds completely, we get the final result:  

`sum_of_digits(568345) = 31`

---

### **Key Points to Remember:**
1. **Breaking down (Forward Process):** Function keeps calling itself until the base case.  
2. **Unwinding (Backward Process):** Function starts returning values step by step.  
3. Think of recursion like stacking plates—first, you stack them up (calls), and then remove them one by one (returns).
