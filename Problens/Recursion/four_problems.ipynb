### **Question 1: Dry Run (Step-by-Step Process)**  
**Given function:**  

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)
```

**Dry run for `sum_of_digits(123)`:**  

#### **Recursive breakdown (Going up):**  
```
sum_of_digits(123) → 3 + sum_of_digits(12)
sum_of_digits(12)  → 2 + sum_of_digits(1)
sum_of_digits(1)   → 1 + sum_of_digits(0)
sum_of_digits(0)   → Base case: return 0
```

#### **Unwinding (Coming down):**  
```
1 + 0 = 1
2 + 1 = 3
3 + 3 = 6
```

**Final output: `6`** ✅

### **Question 2: Identify Base Case**  
**Given function:**  

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

**Base case:**  
```python
if n == 1:
    return 1
```

**Explanation:**  
The base case is the condition that stops the recursion when `n` becomes `1`. Without a base case, the recursion would continue indefinitely and result in a stack overflow error.  

**Answer: `if n == 1: return 1`** ✅

### **Question 3: Predict Output**  
**Given function:**  

```python
def mystery(n):
    if n == 0:
        return 1
    return 2 * mystery(n - 1)

print(mystery(3))
```

**Recursive breakdown:**  

```
mystery(3) → 2 * mystery(2)
mystery(2) → 2 * mystery(1)
mystery(1) → 2 * mystery(0)
mystery(0) → Base case: return 1
```

#### **Unwinding (Coming down):**  
```
2 * 1 = 2
2 * 2 = 4
2 * 4 = 8
```

**Final output: `8`** ✅

### **Question 4: Conceptual Understanding**  
**Question:** What happens during **unwinding** in recursion?  

**Answer:**  
Unwinding is the process where the function starts returning values step-by-step **after reaching the base case**, solving smaller problems first and combining results as it moves back up the recursive calls.  

**Example:**  
For `sum_of_digits(123)`, the calls happen like:  

```
sum_of_digits(123) → sum_of_digits(12) → sum_of_digits(1) → sum_of_digits(0)
```

Once the base case is reached, unwinding happens like:  

```
sum_of_digits(0) → sum_of_digits(1) → sum_of_digits(12) → sum_of_digits(123)
```

Each step returns a value to the previous call until the final result is obtained.  

**Answer: "Unwinding means returning values step-by-step after reaching the base case."** ✅
