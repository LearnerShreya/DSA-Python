#### Data Structures: Arrays  
**Big O of Common Array Operations**

1. **Access**: O(1)  
   - Directly access an element using its index.  

2. **Set**: O(1)  
   - Update the value of an element at a specific index.  

3. **Traverse/Search**: O(n)  
   - Traverse the array or search for an element.  

4. **Copy**: O(n)  
   - Copy all elements from one array to another.  

5. **Insert**:  
   - At the beginning: O(n) (requires shifting all elements).  
   - At the end: O(1) (if space allows, amortized for dynamic arrays).  
   - Somewhere in between: O(n) (requires shifting elements).  

6. **Remove**:  
   - At the beginning: O(n) (requires shifting all elements).  
   - At the end: O(1).  
   - Somewhere in between: O(n) (requires shifting elements).
