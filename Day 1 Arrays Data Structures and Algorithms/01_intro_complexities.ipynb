More formally, moving clockwise from the ith friend brings you to the (i+1)th friend for 1 <= i < n, and moving clockwise from the nth friend brings you to the 1st friend.  
The rules of the game are as follows:  
1.Start at the 1st friend.  
2.Count the next k friends in the clockwise direction including the friend you started at. The counting wraps around the circle and may count some friends more than once.  
3. The last friend you counted leaves the circle and loses the game.  
4.If there is still more than one friend in the circle, go back to step 2 starting from the friend immediately clockwise of the friend who just lost and repeat.  
5.Else, the last friend in the circle wins the game.  
Given the number of friends, n, and an integer k, return the winner of the game.

**Time Complexity** measures the amount of time an algorithm takes to execute as a function of the size of its input. It is an essential aspect of analyzing an algorithm's efficiency and is expressed using **Big-O notation** to describe the worst-case scenario.

### Key Points:
1. **Definition**:  
   Time complexity quantifies the number of basic operations performed by an algorithm relative to the input size (n). It ignores machine-dependent constants and focuses on growth rates.

2. **Importance**:  
   - Helps evaluate the scalability of an algorithm.  
   - Facilitates comparison between algorithms to choose the most efficient one for a task.

3. **Common Types of Time Complexity**:
   - **O(1) (Constant Time)**: Independent of input size (e.g., accessing an array element).
   - **O(log n) (Logarithmic Time)**: Operations reduce the problem size by a constant factor (e.g., binary search).
   - **O(n) (Linear Time)**: Directly proportional to input size (e.g., iterating through an array).
   - **O(n log n)**: More complex than linear but less than quadratic (e.g., merge sort).
   - **O(n²) (Quadratic Time)**: Nested loops over input (e.g., bubble sort).
   - **O(2ⁿ) (Exponential Time)**: Rapidly increases with input size (e.g., brute force solutions for the Travelling Salesman Problem).
   - **O(n!) (Factorial Time)**: Worst growth rate, generally impractical (e.g., solving the N-Queens problem using backtracking).

4. **Factors Influencing Time Complexity**:
   - Input size (n).
   - Algorithm design and implementation.
   - Type of operations performed (loops, recursion, etc.).

5. **How to Analyze Time Complexity**:
   - **Count operations**: Approximate the number of iterations or recursive calls.
   - **Simplify using Big-O**: Focus on the dominant term and discard constants.

6. **Use Cases**:
   Time complexity analysis is vital in selecting algorithms for large-scale data processing, optimizing software, and ensuring responsiveness in real-time systems.

---
Asymptotic analysis is a mathematical approach to evaluate the performance of an algorithm based on its growth behavior as the input size (n) approaches infinity. It helps compare the efficiency of algorithms by analyzing their **time complexity** or **space complexity** independent of hardware or software limitations.

---

### **Purpose**:
- **Compare Algorithms**: Determine which algorithm scales better for large input sizes.
- **Focus on Efficiency**: Identify bottlenecks for optimization.
- **Machine Independence**: Provide insights irrespective of system hardware or compiler optimizations.

---

### **Key Elements**:
1. **Input Size (n)**: The size of the data on which the algorithm operates.
2. **Basic Operations**: Fundamental computations (comparisons, additions, assignments) that dominate execution time.
3. **Growth Rate**: Asymptotic analysis studies how the algorithm's execution time or space requirement grows as n increases.

---

### **Asymptotic Notations**:
1. **Big-O Notation (O)**:
   - Represents the **upper bound** of an algorithm's running time.
   - Describes the **worst-case scenario**.
   - Example: A loop iterating n times has O(n) complexity.

2. **Omega Notation (Ω)**:
   - Represents the **lower bound** of an algorithm's running time.
   - Describes the **best-case scenario**.
   - Example: A binary search algorithm requires at least log n operations, so it is Ω(log n).

3. **Theta Notation (Θ)**:
   - Represents the **tight bound** (both upper and lower bounds are the same).
   - Describes the **average-case scenario** or exact growth rate.
   - Example: An algorithm with c ⋅ n steps is Θ(n).

---

### **Examples**:
**Linear Search**:  
   - Time Complexity:  
     - Worst-case: O(n).  
     - Best-case: Ω(1).  
     - Tight-bound: Θ(n).


---

### **Steps in Asymptotic Analysis**:
1. **Identify Input Size**: Determine the problem size (n).
2. **Count Operations**: Focus on dominant operations (loops, recursive calls).
3. **Ignore Constants**: Discard lower-order terms and constants.
4. **Express Growth**: Use appropriate asymptotic notation.

### 4. What is Big O?  
**Big O Notation (O)** is a mathematical concept used to describe the **upper bound** of an algorithm's running time or space requirements in the worst-case scenario. It helps measure the efficiency of an algorithm by characterizing how its performance (time or space) scales with the size of the input data (\(n\)).

#### **Key Points about Big O**:
1. **Worst-Case Scenario**: Big O focuses on the maximum amount of time or space the algorithm will require for a given input size.
2. **Asymptotic Analysis**: It provides a way to analyze how an algorithm behaves as the input size increases, ignoring constant factors and lower-order terms.
3. **Upper Bound**: Big O represents an upper bound, meaning it gives the worst-case performance that the algorithm will not exceed.

### 5. Common Complexities  

1. **O(1) – Constant Time**:
   - **Definition**: The algorithm’s performance is constant, meaning it takes the same amount of time regardless of the size of the input.
   - **Example**: Accessing an element in an array by index.

2. **O(log n) – Logarithmic Time**:
   - **Definition**: The algorithm’s performance grows logarithmically with the input size. It increases slowly as n grows.
   - **Example**: Binary search on a sorted array.

3. **O(n) – Linear Time**:
   - **Definition**: The algorithm’s performance grows linearly with the input size. If n doubles, the time taken also doubles.
   - **Example**: Scanning through a list of n elements.

4. **O(n log n) – Linearithmic Time**:
   - **Definition**: A combination of linear and logarithmic growth, this is typically seen in efficient sorting algorithms.
   - **Example**: Merge sort or quicksort.

5. **O(n²) – Quadratic Time**:
   - **Definition**: The algorithm’s performance grows quadratically with the input size. It increases much faster than linear time.
   - **Example**: Nested loops (e.g., bubble sort or insertion sort).

6. **O(2^n) – Exponential Time**:
   - **Definition**: The algorithm’s performance grows exponentially with the input size. Even small increases in n lead to massive increases in runtime.
   - **Example**: Solving the traveling salesman problem via brute force.

7. **O(n!) – Factorial Time**:
   - **Definition**: The algorithm’s performance grows factorially with the input size. This is one of the slowest-growing complexities, making it impractical for large n.
   - **Example**: Generating all permutations of n items.

**Definition:**  
Space complexity refers to the total amount of memory or storage an algorithm requires to run to completion, including:

1. **Fixed Part:** The space required for constants, variables, and program instructions.  
2. **Variable Part:** The space required dynamically during execution, such as for recursion, function calls, or data structures.

---

#### **Components of Space Complexity**
1. **Instruction Space:** Memory used to store the compiled instructions.  
2. **Data Space:** Memory required for variables and constants.  
3. **Auxiliary Space:** Temporary or additional memory used during execution (e.g., recursion stack or temporary arrays).  
4. **Input Space:** Memory used to store the input data.

---

#### **Common Space Complexities**
1. **O(1) – Constant Space:**  
   - The algorithm uses a fixed amount of memory, regardless of input size.  
   - **Example:** Swapping two numbers or finding the minimum in a list.

2. **O(n) – Linear Space:**  
   - Memory usage grows linearly with input size.  
   - **Example:** Storing an array or performing Depth-First Search (DFS).

3. **O(n²) – Quadratic Space:**  
   - Memory usage grows quadratically, such as when using a 2D matrix.  
   - **Example:** Storing the adjacency matrix of a graph.

---

#### **Examples**
1. **Iterative Algorithm (O(1) Space):**  
   ```python
   def sum_array(arr):
       total = 0  # Fixed variable
       for num in arr:
           total += num
       return total
   ```
   **Space Complexity:** Constant, since only the variable `total` is used.

2. **Recursive Algorithm (O(n) Space):**  
   ```python
   def factorial(n):
       if n == 0:
           return 1
       return n * factorial(n - 1)
   ```
   **Space Complexity:** Linear, due to the recursion stack growing proportionally to \(n\).

3. **Dynamic Programming (O(n²) Space):**  
   ```python
   def matrix_chain_order(p):
       n = len(p) - 1
       dp = [[0] * n for _ in range(n)]  # 2D matrix
       # Some logic here
       return dp
   ```
   **Space Complexity:** Quadratic, due to the 2D table.

---

#### **Optimizing Space Complexity**
1. **In-place Algorithms:** Minimize space usage by modifying input data directly.  
   - **Example:** Sorting an array in-place (e.g., QuickSort).  

2. **Iterative Approaches:** Replace recursion with iteration to avoid stack overhead.  
   - **Example:** Use a stack or queue explicitly instead of recursive DFS.

3. **Memory-Efficient Data Structures:** Use efficient data structures like heaps, tries, or hash tables to save space.  

---

#### **Why Space Complexity Matters**
1. Critical for memory-constrained environments, like embedded systems.  
2. Ensures scalability for large datasets.  
3. Improves overall performance by reducing memory usage and potential paging.

### 7. Techniques to Simplifying Big O Expressions?  
- drop constant
- drop insignificant terms (as N increases)
- do not drop diff input parameter
  - e.g.: O(N<sup>3</sup>+M)

---
Simplifying Big O expressions helps focus on the dominant factors that influence the growth rate of an algorithm. Here are key techniques for simplification:

**1. Eliminate Constants**  
Big O focuses on the asymptotic behavior, so constant factors are irrelevant as \(n\) grows large.

- Example:  
  O(3n) → O(n)  
  O(5n² + 7) → O(n²)  



**2. Drop Non-Dominant Terms**  
When multiple terms are present, keep only the term with the fastest growth rate.

- Example:  
  O(n² + n) → O(n²)  
  O(n + log n) → O(n)  


**Summary**  
Simplifying Big O ensures clarity and focus on the most influential parts of an algorithm's complexity. It emphasizes scalability, which is critical for analyzing performance as input size grows.

### 8. Logarithms

Logarithms often appear in the time complexity of algorithms, especially those involving divide-and-conquer strategies or binary operations. Here’s a breakdown:

---

#### Definition
- A logarithm is the inverse of exponentiation.  
  logₐ(x) = y means aᵧ = x, where a is the base.  
  In computer science, the base is typically 2, so log(n) usually refers to log₂(n).

---

#### Properties of Logarithms
1. log(ab) = log(a) + log(b)  
2. log(aᵇ) = b * log(a)  
3. log(a/b) = log(a) - log(b)  
4. Change of base: logₐ(b) = logₖ(b) / logₖ(a)

---

#### Common Logarithmic Complexities
1. **O(log n)**:  
   Algorithms that repeatedly divide the input in half.  
   - Examples: Binary search, height of a balanced binary tree.

2. **O(n log n)**:  
   Algorithms that perform log n operations for each of the n elements.  
   - Examples: Merge sort, heap sort.

---

#### Why are Logarithms Common?
- Logarithms arise when the problem size is reduced by a constant factor in each step, such as in binary search.  
- They indicate efficient growth, as log(n) grows much slower than linear or polynomial functions.
