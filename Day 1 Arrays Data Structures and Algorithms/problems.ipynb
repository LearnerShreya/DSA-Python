````python
def monotonic_array(array):
    #write code here
    if len(array) == 0: return True
    first = array[0]
    last = array[len(array)-1]
    
    if first > last:
        for i in range(len(array)-1):
            if array[i]<=array[i+1]:
                return False
        
    elif first == last:
        for i in range(len(array)-1):
            if array[i]!=array[i+1]:
                return False
                
    else:
        for i in range(len(array)-1):
            if array[i]>=array[i+1]:
                return False
                
    return True
````

````python
class Solution:
    def isMonotonic(self, nums):
        """
        Check if the given array is monotonic.
        :type nums: List[int]
        :rtype: bool
        """
        # Start with flags assuming array can be both increasing and decreasing
        is_increasing = True
        is_decreasing = True

        # Traverse the array
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                is_increasing = False
            if nums[i] < nums[i + 1]:
                is_decreasing = False

        # Return True if any one condition holds
        return is_increasing or is_decreasing

sol = Solution()
print(sol.isMonotonic([1, 2, 2, 3]))  # Expected: True
print(sol.isMonotonic([6, 5, 4, 4]))  # Expected: True
print(sol.isMonotonic([1, 3, 2]))     # Expected: False
````

### **Flag Kya Hota Hai?**
Flag ek **variable** hota hai jo sirf **True** ya **False** value rakhta hai.  
- **Purpose**: Yeh hume yeh batata hai ki kuch condition satisfy ho rahi hai ya nahi.

For example:  
- **is_increasing**: Agar array monotonic increasing hai, toh yeh `True` hoga. Agar array increasing na ho, toh yeh `False` kar denge.
- **is_decreasing**: Agar array monotonic decreasing hai, toh yeh `True` hoga. Agar array decreasing na ho, toh yeh `False` kar denge.

Flags ka use karke hum program ka flow control karte hain.

---

### **Ek Simple Example**:

Imagine karo tumhare paas ek array hai: `[1, 2, 3, 4]`.  
- Tum check karna chahte ho ki yeh array increasing hai ya nahi.  

#### Without Flag:
Har element ko compare karte raho. Agar kahin condition fail ho gayi, toh kuch extra logic lagana padega.

#### With Flag:
- Start karte time `is_increasing = True`.  
- Jaise hi koi condition fail kare (e.g., `nums[i] > nums[i+1]`), `is_increasing = False` kar do.  
- End mein bas yeh check karo ki flag `True` hai ya `False`.  

---

### **Simplified Code Without Jargon**:

Abhi code ko step by step breakdown karta hoon:

#### Input Array: `[1, 2, 3, 3, 4]`

1. **Start with Flags**:  
   - `is_increasing = True` (Maan ke chalte hain ki array increasing hai).  
   - `is_decreasing = True` (Maan ke chalte hain ki array decreasing hai).

2. **Loop Through Array**:  
   Har element ko next element ke saath compare karo:  
   - Agar `nums[i] > nums[i+1]`, iska matlab increasing fail ho gaya, toh `is_increasing = False` kar do.  
   - Agar `nums[i] < nums[i+1]`, iska matlab decreasing fail ho gaya, toh `is_decreasing = False` kar do.

3. **Final Check**:  
   - Agar `is_increasing` ya `is_decreasing` me se koi bhi `True` hai, toh array monotonic hai.  
   - Agar dono `False` ho gaye, toh array monotonic nahi hai.

### **Final Code (Bilkul Simple)**:

```python
class Solution(object):
    def isMonotonic(self, nums):
        # Step 1: Start with both flags as True
        is_increasing = True
        is_decreasing = True

        # Step 2: Traverse the array
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:  # Increasing condition fails
                is_increasing = False
            if nums[i] < nums[i + 1]:  # Decreasing condition fails
                is_decreasing = False

        # Step 3: Return the result
        return is_increasing or is_decreasing
```

### **Explain with Example**:

#### Input: `[1, 2, 2, 3]`  
1. Start: `is_increasing = True`, `is_decreasing = True`.  
2. Compare:  
   - `1 <= 2`: Increasing okay.  
   - `2 <= 2`: Increasing okay.  
   - `2 <= 3`: Increasing okay.  
3. Final Flags:  
   - `is_increasing = True`, `is_decreasing = False`.  
4. Result: `True` (Array is monotonic).

#### Input: `[5, 4, 4, 3]`  
1. Start: `is_increasing = True`, `is_decreasing = True`.  
2. Compare:  
   - `5 >= 4`: Decreasing okay.  
   - `4 >= 4`: Decreasing okay.  
   - `4 >= 3`: Decreasing okay.  
3. Final Flags:  
   - `is_increasing = False`, `is_decreasing = True`.  
4. Result: `True` (Array is monotonic).

#### Input: `[1, 3, 2]`  
1. Start: `is_increasing = True`, `is_decreasing = True`.  
2. Compare:  
   - `1 <= 3`: Increasing okay.  
   - `3 > 2`: Increasing fails, decreasing fails.  
3. Final Flags:  
   - `is_increasing = False`, `is_decreasing = False`.  
4. Result: `False` (Array is not monotonic).
